```python
#   

bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]

#   
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

###1-1.  

```python
import matplotlib.pyplot as plt

#   
plt.scatter(bream_length, bream_weight)
plt.title('Bream Data')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()

print()

#   
plt.scatter(smelt_length, smelt_weight)
plt.title('Smelt Data')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

###2.   

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
#print(length)
#print(weight)

fish_data = [[l, w] for l, w in zip(length, weight)]
#print(fish_data)

bream_cnt = len(bream_length)
smelt_cnt = len(smelt_length)
# print(bream_cnt, smelt_cnt)

labels = ['smelt', 'bream']

fish_target = [labels.index('bream')] * bream_cnt + [labels.index('smelt')] * smelt_cnt
# print(fish_target)
```

###3.    
- fit() :  
- score() :  

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier() # n_neigbors = 5
kn.fit(fish_data, fish_target)
kn.score(fish_data, fish_target)

print(kn._fit_X)
print(kn._y)
```

###4.  

```python
# , , Test  

test_data = [30, 600]

plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.scatter(test_data[0], test_data[1], marker='^')

plt.legend(['bream', 'smelt', 'test data'])

plt.title('All Data')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()

prediction_result = kn.predict([test_data])

# 
print(f'test data model prediction result : {labels[int(prediction_result[0])]}')


```
